```python
from collections import defaultdict
d = defaultdict(int)   # é»˜è®¤å€¼å·¥å‚æ˜¯ int
# ä¸€ä¸ªå¯è°ƒç”¨å¯¹è±¡ï¼ˆå‡½æ•°ã€ç±»å‹ç­‰ï¼‰ï¼Œç”¨æ¥ç”Ÿæˆâ€œé»˜è®¤å€¼â€ã€‚
d.default_factory = lambda: 'N/A'
d['a']
d['b']
d
```

```python
# è®¡æ•°å™¨ï¼šdefaultdict(int
counter = defaultdict(int)  # int() -> 0
data = ["a", "b", "a", "c", "b", "a"]

for item in data:
    counter[item] += 1
counter
```

```python
# åˆ†ç»„ï¼š
# å…¸å‹åœºæ™¯ï¼šæ ¹æ®æŸä¸ªå­—æ®µåˆ†ç»„ï¼ˆå¾ˆå¸¸ç”¨ï¼‰ã€‚

students = [
    ("å¼ ä¸‰", "ä¸€ç­"),
    ("æå››", "ä¸€ç­"),
    ("ç‹äº”", "äºŒç­"),
]

grouped = defaultdict(list)
for name, clazz in students:
    grouped[clazz].append(name)
grouped
```

```python
# å»é‡åˆ†ç»„ï¼šdefaultdict(set) ğŸ§¹
# æ¯”å¦‚â€œæ¯ä¸ªæ ‡ç­¾ä¸‹æœ‰å“ªå‡ ä¸ªä¸åŒçš„ç”¨æˆ·â€ã€‚
tags = [
    ("ç”¨æˆ·1", "Python"),
    ("ç”¨æˆ·2", "Python"),
    ("ç”¨æˆ·1", "SQL"),
    ("ç”¨æˆ·2", "Python"),  # é‡å¤
]

tag_users = defaultdict(set)
for user, tag in tags:
    tag_users[tag].add(user)
tag_users
```
