```python
a="hello's world"
print(a)

#转义
a="hello\"s \tworld"
print(a)

# 原始字符串
a=r"hello\"s \tworld"
print(a)

# 字符串拼接(自动拼接)
a = "hello" 'of' "world"
print(a)


a="a\0b\0c"
print(a)

```

```python
# 原始字符串阻止转义
b = r"c:\new\path"
# 'c:\\new\\path'
b
# c:\new\path
print(b)

r = r"c:\new\path\" #错误
r = r"c:\new\path\\" #正确
# 注意：使用r阻止原始字符串转义时，不能以单个反斜杠结尾，因为它会转义引号字符。
```
