```python
from dataclasses import dataclass


class OrderCreated:
    order_id: str
    amount: float
    tags: list[str] = []

    def __init__(self, order_id: str, amount: float) -> None:
        self.order_id = order_id
        self.amount = amount


o1 = OrderCreated("123", 100.0)
o1.tags.append("tag1")
o1.tags.append("tag2")

o2 = OrderCreated("123", 100.0)
print(o2.tags)
```
