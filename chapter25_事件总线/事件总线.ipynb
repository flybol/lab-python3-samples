```python
from collections import defaultdict
from typing import Callable, Dict, List, Type


class Event:
    """æ‰€æœ‰äº‹ä»¶çš„åŸºç±»"""


class EventBus:
    """äº‹ä»¶æ€»çº¿"""

    def __init__(self):
        self._handlers: Dict[Type[Event], List[Callable]] = defaultdict(list)

    def subscribe(self, event_type: Type[Event], listener: Callable):
        """æ³¨å†Œç›‘å¬å™¨"""
        self._handlers[event_type].append(listener)

    def publish(self, event: Event):
        for handler in self._handlers[type(event)]:
            handler(event)


# å®šä¹‰äº‹ä»¶
class OrderCreated(Event):
    def __init__(self, order_id: str):
        self.order_id = order_id


# 3ï¸âƒ£ å®šä¹‰ç›‘å¬è€…ï¼ˆè®¢é˜…è€…ï¼‰


def notify_user(event: OrderCreated):
    print(f"ğŸ“© é€šçŸ¥ç”¨æˆ·ï¼šè®¢å• {event.order_id} å·²åˆ›å»º")


def deduct_stock(event: OrderCreated):
    print(f"ğŸ“¦ æ‰£å‡åº“å­˜ï¼šè®¢å• {event.order_id}")


# ä½¿ç”¨äº‹ä»¶
bus = EventBus()

bus.subscribe(OrderCreated, notify_user)
bus.subscribe(OrderCreated, deduct_stock)


# å‘å¸ƒäº‹ä»¶
bus.publish(OrderCreated("ORDER-1001"))
```
