笔记：参数名可以在调用时与调用者作用域中的变量实现共享对象。因此，在函数中原位置修改可变对象
参数能够影响调用者。

```python
#当参数被传入可变对象
def fn1(a,b):
    a = 2
    b[0] = "hello"

X = 1
# 可变对象，传递的是一个地址，
L = [1,2]
fn1(X,L)
print(X,L)
```

```python
# 参数匹配基础
def func(name,*args):
    # 把不能匹配的基于位置的参数收集到一个元组中
    print(name,args)


func("张三",1,2,3,4)
```

```python
def func1(path:str,**kwargs):
    print(path,kwargs)

#把多出来不能匹配的关键字参数收集到一个字典中
func1("a.py",a=1,b=2,c=3)
```

```python
def fn(value,lt:list|None=None):
    if lt is None: lt = []
    lt.append(value)
    return lt

print(fn(1))
print(fn(2))
print(fn(3))

```
