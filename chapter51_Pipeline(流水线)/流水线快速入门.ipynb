```python
from typing import Callable, Any

Command = Callable[[Any], Any]


##Á¨¨ 1 Èò∂ÊÆµÔºöÂáΩÊï∞Á∫ß PipelineÔºàÊúÄÁÆÄÂçïÁâàÊú¨Ôºâ‚úÖ
class Pipeline:
    def __init__(self, steps: list[Command]) -> None:
        self.steps = steps

    def run(self, data: Any):
        results = [step(data) for step in self.steps]
        return results


def step1(x):
    return x + 1


def step2(x):
    return x * 10


pipeline = Pipeline([step1, step2])
print(pipeline.run(1))
```

```python
from typing import Protocol, runtime_checkable

"""
Á¨¨ 2 Èò∂ÊÆµÔºöÂëΩ‰ª§ÂØπË±°ÂåñÔºàCommand PatternÔºâ‚≠ê
ÂáΩÊï∞‰∏çÂ§üÁî®‰∫ÜÊÄé‰πàÂäûÔºü

ÈúÄË¶ÅÁä∂ÊÄÅ

ÈúÄË¶ÅÈÖçÁΩÆ

ÈúÄË¶ÅÁîüÂëΩÂë®Êúü

üëâ Áî®Á±ª
"""


# 1Ô∏è‚É£ ÂÆö‰πâ Command Êé•Âè£ÔºàProtocolÔºâ
@runtime_checkable
class Command(Protocol):
    async def run(self, data: Any) -> Any: ...


# 2Ô∏è‚É£ ÂÖ∑‰ΩìÂëΩ‰ª§ÂÆûÁé∞
class FetchCommand:
    async def run(self, url: str) -> str:
        return f"<html>{url}</html>"


class ParseCommand:
    async def run(self, html: str) -> dict:
        return {"title": "demo"}


# 3Ô∏è‚É£ ÂºÇÊ≠• Pipeline
class AsyncPipeline:
    def __init__(self, steps: list[Command]):
        self.steps = steps

    async def run(self, data: Any):
        for step in self.steps:
            data = await step.run(data)
        return data


async def main():
    pipeline = AsyncPipeline([FetchCommand(), ParseCommand()])
    result = await pipeline.run("https://example.com")
    print(result)


await main()
```
