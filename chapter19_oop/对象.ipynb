```python
class Person:
    def __init__(self, name, job=None, pay=0):
        self.name = name
        self.job = job
        self.pay = pay

    def giveRaise(self, percent):
        self.pay = int(self.pay * (1.0 + percent))
        return self.pay

    def __repr__(self):
        return f"{self.name} job{self.job} {self.pay:.2f}"


class Manager(Person):
    def __init__(self, name, pay=0):
        Person.__init__(self, name, "mgr", pay)

    # 继承父类的构造函数
    def giveRaise(self, percent, bonus=0.1):
        # self.pay = int(self.pay * (1.0 + percent + bonus))
        # 复用父类方法：关键点是传递当前实例
        Person.giveRaise(self, percent + bonus)
        return self.pay

    def __getattr__(self, name):
        print("getattr:", name)
        return getattr(self, name)


m1 = Manager("Tom", 5000)
# print(m1.giveRaise(0.2))
m1.giveRaise(0.2)
print(m1.name)

for key in m1.__dict__:
    # 访问属性的内容
    print(getattr(m1, key))

# print(Manager.__mro__)
# print(Manager.__bases__)
```

```python
class AttrDisplay:
    """显示对象属性"""

    def gatherAttrs(self):
        attrs = []
        for key in sorted(self.__dict__):
            attrs.append("%s=%s" % (key, getattr(self, key)))
        return ", ".join(attrs)

    def __repr__(self):
        return "[%s: %s]" % (self.__class__.__name__, self.gatherAttrs())


class Person(AttrDisplay):
    def __init__(self, name, job=None, pay=0):
        self.name = name
        self.job = job
        self.pay = pay


p = Person("Bob", "dev", 5000)
print(p)
```

```python
# 验证实例是否会从类中找属性？
class A:
    a: int = 100


# 所有实例共享的属性
a1 = A()
print(a1.a)

a2 = A()
print(a2.a)

# 陷阱：以为是在修改类的属性，其实是新增实例的私有属性。
a2.a = 200
print(a2.a)
print(a1.a)

# 说明所有实例都共享类的属性
a3 = A()
print(a3.a)


# 修改类的属性值
a = A
a.a = 300

print(a1.a)
# 先访问实例的属性
print(a2.a)

print(a3.a)
```

```python
class Super:
    def __init__(self, a):
        self.a = a
        print("super init")


class Sub(Super):
    def __init__(self, a, b):
        super().__init__(a)
        self.b = b
        print("sub init")


sub = Sub(1, 2)
print(sub.__dict__)
```

```python
# 子类覆盖父类的方法
class Super:
    def fnA(self):
        print("I am super")


class Sub(Super):
    def fnA(self):
        # 复用父类的方法
        super().fnA()
        # 扩展父类方法
        print("I am sub")


subb = Sub()
subb.fnA()
```

```python
def classtree(cls, indent=4):
    print("." * indent + cls.__name__)
    for supercls in cls.__bases__:
        classtree(supercls, indent + 4)


def instancetree(inst):
    print("Tree of %s" % inst)
    classtree(inst.__class__, 4)


def selftest():
    class A:
        pass

    class B(A):
        pass

    class C(A):
        pass

    class D(B, C):
        pass

    class E:
        pass

    class F(D, E):
        pass

    instancetree(B())
    instancetree(F())


selftest()
```

```python
class Iters:
    def __init__(self, value):
        self.data = value

    def __getitem__(self, i):
        print("getitem", i, end=" ")
        return self.data[i]

    def __iter__(self):
        print("iter", end=" ")
        self.ix = 0
        return self

    def __next__(self):
        print("next", end=" ")
        if self.ix == len(self.data):
            raise StopIteration
        item = self.data[self.ix]
        self.ix += 1
        return item

    def __contains__(self, x):
        print("contains", x, end=" ")
        return x in self.data

    next = __next__


X = Iters([1, 2, 3, 4, 5])
# print(1 in X)


# for x in X:
#     print(x)

# print([i**2 for i in X])
# print(list(map(lambda x: x**2, X)))

X[1:3]
```
