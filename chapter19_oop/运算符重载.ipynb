```python
class A:
    def __init__(self, x: int):
        self.x = x

    def __add__(self, other):
        return self.x + other.x

    def __sub__(self, other):
        return self.x - other.x


a = A(10) + A(20)
print(a)
b = A(10) - A(20)
print(b)
```

```python
# 自定义可迭代对象
class Fib:
    def __init__(self, start, stop):
        self.start = start
        self.stop = stop

    def __next__(self):
        # 返回遍历元素的值
        if self.start >= self.stop:
            raise StopIteration
        else:
            # 索引
            self.start += 1
            return self.start - 1

    def __iter__(self):
        return self


for i in Fib(1, 10):
    print(i)
```
